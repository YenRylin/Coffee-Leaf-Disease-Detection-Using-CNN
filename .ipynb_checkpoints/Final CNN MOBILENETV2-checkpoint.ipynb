```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
# Set parameters
BATCH_SIZE = 32
IMAGE_SIZE = 224
EPOCHS = 8  # Tambahkan epochs untuk pelatihan lebih lama
CHANNELS = 3
N_CLASSES = 4
```

```python
# Normalisasi dataset (normalisasi ke rentang 0-1)
def normalize_img(image, label):
    image = tf.cast(image, tf.float32) / 255.0
    return image, label
```

```python
# Load dataset
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "./data/train",
    seed=123,
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
).map(normalize_img)

val_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "./data/valid",
    seed=123,
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
).map(normalize_img)

test_ds = tf.keras.preprocessing.image_dataset_from_directory(
    "./data/test",
    seed=123,
    shuffle=True,
    image_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=BATCH_SIZE
).map(normalize_img)
```

```python
# Prefetch data
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.prefetch(buffer_size=AUTOTUNE)
```

```python
# Load MobileNetV2 base model
base_model = tf.keras.applications.MobileNetV2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
)

# Freeze base model
base_model.trainable = False

# Build model
model = Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.BatchNormalization(),
    layers.Dense(64, activation='relu'),  # Tingkatkan jumlah unit
    layers.Dropout(0.5),  # Dropout lebih besar untuk regularisasi
    layers.Dense(N_CLASSES, activation='softmax')
])
```

```python
# Compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),  # Tingkatkan learning rate awal
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from tensorflow.keras.callbacks import LearningRateScheduler

# Fungsi scheduler untuk mengurangi learning rate berdasarkan epoch
def scheduler(epoch, lr):
    if epoch < 3:
        return lr  # Tetap sama di awal
    else:
        return float(lr * tf.math.exp(-0.1))  # Ubah hasil ke tipe float
```

```python
# Callbacks
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=1, min_lr=1e-6)
stop_early = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = LearningRateScheduler(scheduler, verbose=1)
```

```python
# Initial training (transfer learning)
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[reduce_lr, stop_early, lr_scheduler]
)

# Fine-tuning
base_model.trainable = True
fine_tune_at = len(base_model.layers) - 100  # Unfreeze 100 layer terakhir
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Re-compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Fine-tuning
fine_tuning_history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS,
    callbacks=[reduce_lr, stop_early]
)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(test_ds)
print(f"Test Accuracy: {accuracy:.2f}")
```

```python
import numpy as np
import tensorflow as tf

# **Validation with intentionally wrong labels**
# Ambil label asli dari dataset validasi
val_images = []
val_labels = []
for img, lbl in val_ds.unbatch():  # Unbatch untuk mengakses gambar dan label
    val_images.append(img)
    val_labels.append(lbl.numpy())

val_images = np.array(val_images)
val_labels = np.array(val_labels)

# Buat label salah (acak urutan label)
wrong_labels = np.random.permutation(val_labels)

# Prediksi ulang menggunakan model terlatih
predicted_probs = model.predict(val_images)
predicted_labels = np.argmax(predicted_probs, axis=1)

# Evaluasi apakah prediksi cocok dengan label asli
print("Validation Results:")
for idx, (img, actual, wrong, pred) in enumerate(zip(val_images, val_labels, wrong_labels, predicted_labels)):
    confidence = np.max(predicted_probs[idx])
    status = "Correct" if pred == actual else "Wrong"
    print(f"Image {idx + 1}: Actual Label = {actual}, Wrong Label = {wrong}, Predicted Label = {pred}, "
          f"Confidence = {confidence:.2f}, Status = {status}")
```

```python
# **Test without labels**
# Ambil gambar dari test set
test_images = []
for img, _ in test_ds.unbatch():  # Ignore the labels
    test_images.append(img)

test_images = np.array(test_images)

# Prediksi menggunakan model terlatih
test_predicted_probs = model.predict(test_images)
test_predicted_labels = np.argmax(test_predicted_probs, axis=1)

# Output hasil prediksi
print("\nTest Results:")
for idx, (img, pred) in enumerate(zip(test_images, test_predicted_labels)):
    confidence = np.max(test_predicted_probs[idx])
    print(f"Image {idx + 1}: Predicted Label = {pred}, Confidence = {confidence:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# **Performance on Validation Set**
print("\n--- Validation Set Performance ---")
# Metrik Evaluasi untuk validasi
validation_accuracy = np.sum(predicted_labels == val_labels) / len(val_labels)
print(f"Validation Accuracy: {validation_accuracy:.2%}")

# Confusion Matrix untuk validasi
val_cm = confusion_matrix(val_labels, predicted_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(val_cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(N_CLASSES), yticklabels=range(N_CLASSES))
plt.title('Confusion Matrix for Validation Set')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report untuk validasi
print("\nClassification Report for Validation Set:")
print(classification_report(val_labels, predicted_labels, target_names=[f"Class {i}" for i in range(N_CLASSES)]))

# **Performance on Test Set**
print("\n--- Test Set Performance ---")
# Prediksi untuk test set
true_labels = []  # Ground truth yang benar (jika tersedia)
for _, lbl in test_ds.unbatch():
    true_labels.append(lbl.numpy())
true_labels = np.array(true_labels)

test_accuracy = np.sum(test_predicted_labels == true_labels) / len(true_labels)
print(f"Test Accuracy: {test_accuracy:.2%}")

# Confusion Matrix untuk test set
test_cm = confusion_matrix(true_labels, test_predicted_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(test_cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(N_CLASSES), yticklabels=range(N_CLASSES))
plt.title('Confusion Matrix for Test Set')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report untuk test set
print("\nClassification Report for Test Set:")
print(classification_report(true_labels, test_predicted_labels, target_names=[f"Class {i}" for i in range(N_CLASSES)]))
```

```python
# Plotting function
def plot_training(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(acc) + 1)

    plt.figure(figsize=(14, 6))

    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label='Training Accuracy', color='blue')
    plt.plot(epochs, val_acc, label='Validation Accuracy', color='orange')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label='Training Loss', color='blue')
    plt.plot(epochs, val_loss, label='Validation Loss', color='orange')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Call the function to plot
plot_training(history)
```
